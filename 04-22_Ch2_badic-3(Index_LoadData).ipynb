#### Indexing Rows, Columns, and Submatrices

```python
A[1] #選出第2個 raw
```

```python
A[[1,3]] #選出 [1,3] 兩個 raw
```

```python
A[:,[0,2]] # 選出第 1, 3 個 column
```

```python
np.array([A[1,0],A[3,2]])
```

```python
A[[1,3],[0,2]] #等同於 np.array([A[1,0],A[3,2]])
```

```python
A[[1,3],[0,2,3]] #兩個 list 長度要一樣
```

```python
A[[1,3]][:,[0,2,3]] #先取 2,4 rows, 再取 1,3 column
```

```python
idx = np.ix_([1,3],[0,2,3])
A[idx]
```

```python
print(A)
A[1:4:2 ,0:3:2] # 1:4:2 (slice) 代表 (start:end:間格)，所以是第 1+1 和第 3+1 row; 0:3:2 為 column 0+1,2+1  
```

#### Boolean (布林) Indexing

Boolean: True or False, 

```python
keep_rows = np.zeros(A.shape[0], bool) # A.shape[0]: array A 的 row 數目,  0 = False 
keep_rows
```

```python
keep_rows[[1,3]] = True #令第2,4 位置為 TRUE
keep_rows
```

```python
np.all(keep_rows == np.array([0,1,0,1])) #檢查 array 的元素是否都是 True
```

```python
A[np.array([0,1,0,1])]
```

```python
A[keep_rows]
```

```python
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
idx_bool = np.ix_(keep_rows , keep_cols) # ix_():對目標 array 選取需要的資料, keep_raw: 1,3; keep_col: 0,2,3
A[idx_bool]
```

```python
idx_mixed = np.ix_([1,3], keep_cols) #ix_()裡面可以放 list or Booleans
```

#### More on Selecting Rows and Columns


```python
idx_80 = Auto_re['year'] > 81
Auto_re.loc[idx_80 , ['weight', 'origin']]
```

```python
Auto_re.loc[lambda df: df['year'] > 81, ['weight', 'origin']] #function lambda: df 為 argument
```

```python
Auto_re.loc[lambda df: (df['year'] > 81) & (df['mpg'] > 30),['weight', 'origin']]
```

```python
Auto_re.loc[lambda df: (df['displacement'] < 300) & (df.index.str.contains('ford') | df.index.str.contains('datsun')),
                        ['weight', 'origin']]
```
